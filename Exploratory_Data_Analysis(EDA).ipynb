```python
import plotly.express as px
```

```python
import plotly.figure_factory as ff
```

```python
sns.pairplot(df, hue='class', markers=["o", "^"], palette=['mediumspringgreen', 'blueviolet']);
```

```python
x0 = np.array(df[df['class']==0]['variance'])
x1 = np.array(df[df['class']==1]['variance'])

group_labels = ['Fake Notes', 'Original Notes']
colors = ['tomato', 'blueviolet']

fig = ff.create_distplot([x0, x1], group_labels, bin_size=.5,
                         curve_type='normal', colors=colors)

fig.update_layout(title_text='Distplot & Normal Distribution of Variance')
fig.show()
```
